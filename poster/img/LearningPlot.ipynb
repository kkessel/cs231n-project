```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set()
```

```python
p_one_selfish = pd.read_csv('./selfish_p_one.csv', header=0, index_col=0)['Value'].values
p_one_coop = pd.read_csv('./coop_p_one.csv', header=0, index_col=0)['Value'].values
p_neg_selfish = pd.read_csv('./selfish_p_neg.csv', header=0, index_col=0)['Value'].values
p_neg_coop = pd.read_csv('./coop_p_neg.csv', header=0, index_col=0)['Value'].values
```

```python
sup = pd.read_csv('./selfish_p_one.csv', header=0, index_col=0)['Step'].values
```

```python
font_size=14
selfish_handle = plt.plot(sup[:600],p_one_selfish[:600], color=sns.color_palette()[2], label="Selfish Player")
coop_handle = plt.plot(sup[:600], p_one_coop[:600], color=sns.color_palette()[1], label="Cooperative Player")

plt.legend(prop={'size': font_size})

plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), labelsize=font_size)
plt.tick_params(labelsize=font_size)
plt.title('Learning Profile', fontsize=font_size)
plt.xlabel(r'Steps', fontsize=font_size)
plt.ylabel(r'Score', fontsize=font_size)
plt.savefig('learning.pdf', format='pdf', dpi=1000, bbox_inches="tight")
plt.show()
```
